```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Plot histogram of Years of Experience (YOE)
plt.hist(
    df["YOE"],
    bins=15,
    edgecolor="black",         
    rwidth=0.9           
)

plt.title("Distribution of Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Number of Candidates")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histogram of Code Challenge Scores
plt.figure(figsize=(8,5))

plt.hist(
    df["Code Challenge Score"],
    bins=10,
    edgecolor="black",
    rwidth=0.9
)

plt.title("Code Challenge Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Candidates")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
